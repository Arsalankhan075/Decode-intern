```python

plt.figure(figsize=(8, 6))


counts = data["Quantity"].value_counts().sort_index()

plt.bar(counts.index, counts.values)
plt.title("Quantity Purchased Per Order")
plt.xlabel("Quantity")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Kis ya mahine mein sab se zyada sales ho rahi hain?
data["Date"]=pd.to_datetime(data["Date"])
monthly_data =data.groupby(data["Date"].dt.strftime("%b"))["TotalPrice"].sum().reset_index()

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(x=monthly_data["Date"], height=monthly_data["TotalPrice"])
plt.xlabel("Month")
plt.ylabel("Total Price")
plt.title("Monthly Sales Chart")
plt.xticks(rotation=45) 
plt.show()


```

```python
data.sample(5)
```

```python
# Konsa Product sab se zyada bik raha hai?
da=data.groupby(data.Product)["Quantity"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6)) 
plt.bar(x=da.index, height=da.values)
plt.xlabel("Product Name")
plt.ylabel("Kul Quantity (Total Sold)")
plt.title("Sab Se Zyada Bikne Wale Products")

plt.xticks(rotation=45,) 
plt.show()
```

```python
# Log sab se zyada konsa Payment Method istemal kar rahe hain?


method_counts = data['PaymentMethod'].value_counts()
payment_methods = method_counts.index
tota_value= method_counts.values
colors = ['#2ecc71', '#f1c40f', '#3498db', '#9b59b6', '#e74c3c', '#e74c3c', '#34495e']
colors = colors[:len(payment_methods)]

# 5. Plot the pie chart safely
plt.figure(figsize=(8, 6))
plt.pie(
    x=tota_value,              
    explode=(0.1,0,0,0,0), 
    labels=payment_methods, 
    colors=colors, 
    autopct='%1.1f%%', 
    shadow=True, 
    startangle=140
)

plt.title('Most Popular Payment Methods Market Share', fontsize=14, fontweight='bold', pad=20)
plt.axis('equal')
plt.show()

```
